```python
car_sales_shuffled
```

```python
# The sample function will come in handy when there is million's of data and you just want to use the fration of it
car_sales_shuffled.sample(frac=0.2) # Only select 20% of data
```

```python
# Resetting the index in shuffled
car_sales_shuffled.reset_index(inplace=True, drop=True) # Using drop we get rid of orginal suffled index
```
